## Context
The data consist of the estimation of obesity levels in people from the countries of Mexico, Peru and Colombia, with ages between 14 and 61 and diverse eating habits and physical condition , data was collected using a web platform with a survey where anonymous users answered each question, then the information was processed obtaining 17 attributes and 2111 records.


The data contains numerical data and continous data, so it can be used for analysis based on algorithms of classification, prediction, segmentation and association. Data is available in CSV format

## Objective

The primary objective of this project is to leverage machine learning techniques to:

* **Predict Obesity Levels:** Develop and evaluate predictive models to accurately classify individuals into different obesity levels (Insufficient Weight, Normal Weight, Overweight, and various Obesity Types) based on their demographic information, eating habits, and physical condition.

* **Identify Key Factors:** Gain insights into the most influential factors (eating habits, physical activity, technology use, etc.) that contribute to higher obesity levels. This knowledge can inform targeted interventions and public health initiatives.

Ultimately, the project aims to contribute to a better understanding of obesity and its associated factors, empowering individuals and healthcare professionals to make informed decisions for improved health outcomes.

## About Dataset

This dataset encompasses data for the estimation of obesity levels in individuals from the countries of Mexico, Peru, and Colombia, based on their eating habits and physical condition. 

**Key Points:**

* **Records:** 2111 
* **Attributes:** 17 
* **Target Variable:** `NObesity` (Obesity Level)
* **Target Classes:** 
    * Insufficient Weight
    * Normal Weight
    * Overweight Level I
    * Overweight Level II
    * Obesity Type I
    * Obesity Type II
    * Obesity Type III

**Attribute Details:**

1. `Gender` 
2. `Age`
3. `Height` 
4. `Weight` 
5. `family_history_with_overweight`
6. `FAVC`: Frequent consumption of high-calorie food 
7. `FCVC`: Frequency of consumption of vegetables 
8. `NCP`: Number of main meals per day 
9. `CAEC`: Consumption of food between meals
10. `SMOKE`: Smoker 
11. `CH2O`: Consumption of water daily 
12. `SCC`: Calories consumption monitoring 
13. `FAF`: Physical activity frequency per week 
14. `TUE`: Time using technology devices per day 
15. `CALC`: Consumption of alcohol 
16. `MTRANS`: Transportation used 
17. `NObesity`: TARGET Target variable 

**Note:**

* The dataset is suitable for various machine learning tasks, including classification, clustering, and regression, due to the mix of numerical and categorical data and the presence of a labeled target variable.

```python
def plot_categorical_features(data, categorical_features):
    # Create an empty dictionary to store the mapping between feature names and their corresponding labels for plotting
    names = {}

    # Initialize a counter variable to 1 to keep track of the subplot positions
    count = 1

    # Iterate through each feature in the categorical_features list
    for feature in categorical_features:

        # Check if the feature is 'ObesityLevel' and if so, continue to the next iteration
        if feature == "ObesityLevel":
            continue

        # Print a separator line, the feature name in bold and underlined format, and another separator line
        print("-" * 150)
        print(f"Feature : \033[4m\033[1m{feature}\033[0m\033[0m")
        print("-" * 150)

        # If the feature is 'Gender_Male', assign the labels 'Male' and 'Female', otherwise get unique values of the feature and convert them to a list to be used as labels
        labels = (
            ["Male", "Female"]
            if feature == "Gender_Male"
            else data[feature].unique().tolist()
        )

        # Create a figure with two subplots arranged in a single row
        fig, axes = plt.subplots(1, 2, figsize=(18, 6))

        # Check if the count is even
        if count % 2 == 0:
            # Plot countplot of feature with respect to target
            sns.countplot(
                x="ObesityLevel", data=data, hue=feature, ax=axes[0], palette="rainbow"
            )

            # Plot pie chart to show distribution of feature
            axes[1].pie(
                data[feature].value_counts().values,
                labels=labels,
                autopct="%1.1f%%",
                startangle=90,
            )
            axes[1].set_xlabel(feature, size=22)

        else:
            # Plot pie chart to show distribution of feature
            axes[0].pie(
                data[feature].value_counts().values,
                labels=labels,
                autopct="%1.1f%%",
                startangle=90,
            )
            axes[0].set_xlabel(feature, size=22)

            # Plot countplot of feature with respect to target
            sns.countplot(
                x="ObesityLevel", data=data, hue=feature, ax=axes[1], palette="rainbow"
            )

        # Increase the counter
        count += 1

        # Save features
        fig.savefig(f"{feature}_feature.png")

        # Show all plots
        plt.show()
```

```python
numerical_cols = ['Age', 'Height', 'Weight', 'BMI','FCVC','NCP','CH2O',	'FAF','TUE']
ordinal_cols = ['CALC', 'CAEC']
binary_cols = ['Gender', 'family_history_with_overweight', 'FAVC', 'SMOKE', 'SCC']
categorical_cols = ['MTRANS']
onehot_cols = binary_cols + categorical_cols
```

```python
data['ObesityLevel'].unique()
```

```python
data['MTRANS'].unique()
```

```python
ordinal = OrdinalEncoder(categories=[['no', 'Sometimes','Frequently', 'Always'],
                                ['no', 'Sometimes','Frequently', 'Always']],
                               handle_unknown = 'use_encoded_value',
                               unknown_value = -1)

onehot=OneHotEncoder(categories=[
                            ['Female', 'Male'], # Gender
                            ['no', 'yes'], # family_history_with_overweight
                            ['no', 'yes'], # FAVC
                            ['no', 'yes'], # SMOKE
                            ['no', 'yes'], # SCC
                            ['Walking', 'Public_Transportation', 'Automobile', 'Motorbike', 'Bike']],# MTRAN
                             handle_unknown='ignore')

#Combining
encoder = ColumnTransformer([
    # List of (name, transformer, columns) tuples specifying the transformer objects to be applied to subsets of the data.
    ('numerical','passthrough',numerical_cols),
    ('ordinal',ordinal,ordinal_cols),
    ('onehot',onehot,onehot_cols)
])

encoder
```

```python
# Define the label mapping dictionary
label_mapping = {
    'Normal_Weight':0,
    'Overweight_Level_I':1, 
    'Overweight_Level_II':2,
    'Obesity_Type_I':3, 
    'Insufficient_Weight':4, 
    'Obesity_Type_II':5,
    'Obesity_Type_III':6
}

# Map the labels to integers
data['ObesityLevel'] = data['ObesityLevel'].map(label_mapping)

print(data['ObesityLevel'].unique())
```

```python
X=data.drop('ObesityLevel',axis=1)
y=data['ObesityLevel']
print(X.shape, y.shape)
```

```python
y
```

```python
X_train,X_test, y_train, y_test=train_test_split(X,y,test_size=0.2, random_state=7, stratify = y)
print(X_train.shape,X_test.shape, y_train.shape, y_test.shape)
```

```python
y_train
```

```python
lr=LogisticRegression()

scalar=StandardScaler()

pipeline_lr=Pipeline([
    ('encoder',encoder),
    ('scalar',scalar),
    ('lr',lr)
])
param_grid = {
    'lr__C': [0.001, 0.01, 0.1, 1, 10, 100],
    'lr__penalty': ['l2'],
    'lr__solver': ['newton-cg', 'lbfgs', 'sag']
}
cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=7)

# GridSearchCV
grid = GridSearchCV(
    pipeline_lr,
    param_grid,
    scoring=['balanced_accuracy', 'f1_weighted'],
    refit='f1_weighted',
    cv=cv,
    n_jobs=-1  # Use all available cores for parallel processing
)

# Fit the model
grid.fit(X_train, y_train)

```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
lr_pred=grid.predict(X_test)
```

```python
cm=confusion_matrix(y_test,lr_pred)
cm
```

```python
result = classification_report(y_test, lr_pred)

# print the result
print(result)
```

```python
rfc = RandomForestClassifier(
    n_estimators = 100,  # default
    max_features='sqrt',
    class_weight='balanced', # ensure balanced weights to account for class imbalance
    random_state=7)

pipe_rfc =  Pipeline([
    ('encoder', encoder),
    ('rfc', rfc)
])

### Parameters to search over ###
param_grid = {
    'rfc__max_depth': [8, 10, 12, 14, 16, 18, 20, 25],
    'rfc__max_leaf_nodes': [50, 75, 100, 125, 150, 200, 250, 300],
    'rfc__min_samples_split': [50, 75, 100, 150, 200]
}

### Validator ###
# Use 5 k-fold, considering the size of the dataset & computational power
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=7)

## GridSearch ###
grid_rf = GridSearchCV(pipe_rfc,       # model to use
                    param_grid, # parameters to search over
                    scoring = ['balanced_accuracy', 'f1_weighted'], # metrics to compute
                    refit = 'f1_weighted',  # which metric to use to decide the best model
                    cv = cv,
                    n_jobs = -1
                   )

grid_rf.fit(X_train, y_train)
```

```python
print(grid_rf.best_params_)
print(grid_rf.best_score_)
```

```python
lr_pred=grid_rf.predict(X_test)
```

```python
cm=confusion_matrix(y_test,lr_pred)
cm
```

```python
result = classification_report(y_test, lr_pred)

# print the result
print(result)
```

```python
xgb_classifier = XGBClassifier(booster='gbtree', random_state=7)  # Adjust num_class if needed

# Create the pipeline with XGBoost
pipeline_xgb = Pipeline([
    ('encoder', encoder),
    ('xgb', xgb_classifier)
])

# Define the parameter grid for XGBoost
param_grid_xgb = {
    'xgb__n_estimators': [50, 100, 200], 
    'xgb__learning_rate': [0.01, 0.1, 0.2],
    'xgb__max_depth': [3, 5, 7],
    'xgb__subsample': [0.8, 1.0],
    'xgb__colsample_bytree': [0.8, 1.0]
}

# GridSearchCV with XGBoost
grid_xgb = GridSearchCV(
    pipeline_xgb,
    param_grid_xgb,
    scoring=['balanced_accuracy', 'f1_weighted'],
    refit='f1_weighted',
    cv=cv,
    n_jobs=-1 
)

# Fit the model with XGBoost
grid_xgb.fit(X_train, y_train)
```

```python
print(grid_xgb.best_score_)
```

```python
xgb_pred=grid_xgb.predict(X_test)
```

```python
result = classification_report(y_test, xgb_pred)

# print the result
print(result)
```
