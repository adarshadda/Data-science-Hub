## Context
The data consist of the estimation of obesity levels in people from the countries of Mexico, Peru and Colombia, with ages between 14 and 61 and diverse eating habits and physical condition , data was collected using a web platform with a survey where anonymous users answered each question, then the information was processed obtaining 17 attributes and 2111 records.


The data contains numerical data and continous data, so it can be used for analysis based on algorithms of classification, prediction, segmentation and association. Data is available in CSV format

## Objective

The primary objective of this project is to leverage machine learning techniques to:

* **Predict Obesity Levels:** Develop and evaluate predictive models to accurately classify individuals into different obesity levels (Insufficient Weight, Normal Weight, Overweight, and various Obesity Types) based on their demographic information, eating habits, and physical condition.

* **Identify Key Factors:** Gain insights into the most influential factors (eating habits, physical activity, technology use, etc.) that contribute to higher obesity levels. This knowledge can inform targeted interventions and public health initiatives.

Ultimately, the project aims to contribute to a better understanding of obesity and its associated factors, empowering individuals and healthcare professionals to make informed decisions for improved health outcomes.

## About Dataset

This dataset encompasses data for the estimation of obesity levels in individuals from the countries of Mexico, Peru, and Colombia, based on their eating habits and physical condition. 

**Key Points:**

* **Records:** 2111 
* **Attributes:** 17 
* **Target Variable:** `NObesity` (Obesity Level)
* **Target Classes:** 
    * Insufficient Weight
    * Normal Weight
    * Overweight Level I
    * Overweight Level II
    * Obesity Type I
    * Obesity Type II
    * Obesity Type III

**Attribute Details:**

1. `Gender` 
2. `Age`
3. `Height` 
4. `Weight` 
5. `family_history_with_overweight`
6. `FAVC`: Frequent consumption of high-calorie food 
7. `FCVC`: Frequency of consumption of vegetables 
8. `NCP`: Number of main meals per day 
9. `CAEC`: Consumption of food between meals
10. `SMOKE`: Smoker 
11. `CH2O`: Consumption of water daily 
12. `SCC`: Calories consumption monitoring 
13. `FAF`: Physical activity frequency per week 
14. `TUE`: Time using technology devices per day 
15. `CALC`: Consumption of alcohol 
16. `MTRANS`: Transportation used 
17. `NObesity`: TARGET Target variable 

**Note:**

* The dataset is suitable for various machine learning tasks, including classification, clustering, and regression, due to the mix of numerical and categorical data and the presence of a labeled target variable.

```python
# Function to plot a numeric feature
def plot_numeric_feature(data, numeric_features):
    # Iterate throw each feature
    for feature in numeric_features:
        print("-"*150)
        print(f"Feature : \033[4m\033[1m{feature}\033[0m\033[0m")
        print("-"*150)
        
        # Create subplots figure
        fig, axes = plt.subplots(1, 2, figsize=(14, 6))
        
        # Plot histogram
        sns.histplot(data=data, x=feature, ax = axes[0])
        
        # Boxplot
        sns.boxplot(y = feature , x = 'ObesityLevel', data = data, ax = axes[1] )
        
        # Displot of given feature with respect to output variable
        sns.displot(data=data, x=feature, hue="ObesityLevel", multiple="stack", kind="kde")
        
        # Save the plot
        fig.savefig(f"{feature}_Feature.png")


        # Show all plots
        plt.show()
```

```python
def plot_categorical_features(data, categorical_features):
    # Create an empty dictionary to store the mapping between feature names and their corresponding labels for plotting
    names = {}

    # Initialize a counter variable to 1 to keep track of the subplot positions
    count = 1

    # Iterate through each feature in the categorical_features list
    for feature in categorical_features:

        # Check if the feature is 'ObesityLevel' and if so, continue to the next iteration
        if feature == "ObesityLevel":
            continue

        # Print a separator line, the feature name in bold and underlined format, and another separator line
        print("-" * 150)
        print(f"Feature : \033[4m\033[1m{feature}\033[0m\033[0m")
        print("-" * 150)

        # If the feature is 'Gender_Male', assign the labels 'Male' and 'Female', otherwise get unique values of the feature and convert them to a list to be used as labels
        labels = (
            ["Male", "Female"]
            if feature == "Gender_Male"
            else data[feature].unique().tolist()
        )

        # Create a figure with two subplots arranged in a single row
        fig, axes = plt.subplots(1, 2, figsize=(18, 6))

        # Check if the count is even
        if count % 2 == 0:
            # Plot countplot of feature with respect to target
            sns.countplot(
                x="ObesityLevel", data=data, hue=feature, ax=axes[0], palette="rainbow"
            )

            # Plot pie chart to show distribution of feature
            axes[1].pie(
                data[feature].value_counts().values,
                labels=labels,
                autopct="%1.1f%%",
                startangle=90,
            )
            axes[1].set_xlabel(feature, size=22)

        else:
            # Plot pie chart to show distribution of feature
            axes[0].pie(
                data[feature].value_counts().values,
                labels=labels,
                autopct="%1.1f%%",
                startangle=90,
            )
            axes[0].set_xlabel(feature, size=22)

            # Plot countplot of feature with respect to target
            sns.countplot(
                x="ObesityLevel", data=data, hue=feature, ax=axes[1], palette="rainbow"
            )

        # Increase the counter
        count += 1

        # Save features
        fig.savefig(f"{feature}_feature.png")

        # Show all plots
        plt.show()
```

```python
categorical_feature = [
    "Gender_Male",
    "family_history_with_overweight",
    "FAVC",
    "CAEC",
    "SMOKE",
    "SCC",
    "CALC",
    "MTRANS_Bike",
    "MTRANS_Motorbike",
    "MTRANS_Public_Transportation",
    "MTRANS_Walking",
    "ObesityLevel",
    
]
```

```python
plot_categorical_features(data,categorical_feature)
```

### Analysis of Categorical Features and Numerical Feature and their Impact on Obesity Levels


| Categorical Features | Analysis |
|---|---|
| Gender and Obesity | The distribution of obesity levels is relatively similar between males and females, suggesting that gender may not be a strong predictor of obesity in this dataset.|
| Family History's Strong Influence | Individuals with a family history of overweight are more likely to have higher obesity levels, highlighting the role of genetics and shared lifestyle factors. |
| Frequent Consumption of High-Calorie Foods (FAVC) | Those frequently consuming high-calorie foods tend to have elevated obesity levels, emphasizing the impact of dietary choices. |
| Frequency of Consumption of Vegetables (FCVC) | Surprisingly, there's a relatively even distribution across obesity levels for those who consume food between meals, suggesting other factors might play a more dominant role.|
| Smoking's Limited Impact | The majority of individuals in the dataset are non-smokers, and smoking doesn't appear to have a strong association with obesity levels in this particular sample.However, it's crucial to remember that smoking has numerous other severe health risks and should not be considered a weight-loss strategy. |
| Physical Activity and Transportation Mode | Individuals relying on public transportation show a higher prevalence of obesity compared to those using active modes like walking or biking, suggesting a link between physical activity levels and obesity. |
| Varying Calorie Consumption at Meals (CALC) | The distribution of obesity levels across different levels of calorie consumption at meals indicates a complex relationship, potentially influenced by other lifestyle and dietary habits. |


|Numerical Feature | Analysis |
|---|---|
| Age | While the average age is around 24, there's a subtle trend of increasing age with higher obesity levels, suggesting age might be a contributing factor. |
| Height | The weak negative correlation and plot indicate a subtle trend of shorter individuals having slightly higher obesity levels, potentially related to other factors like body composition. |
| Weight | The strong positive correlation and clear upward trend in the plot confirm that weight is a major determinant of obesity. |
| Frequency of Consumption of Vegetables (FCVC) | The moderate negative correlation and downward trend highlight the potential protective effect of vegetable consumption against obesity. |
| Number of Main Meals (NCP) | The weak positive correlation and relatively flat trend suggest that the number of main meals might not have a substantial independent impact on obesity levels. |
| Water Consumption (CH2O) | The weak positive correlation and slight upward trend hint at a potential association between higher water intake and higher obesity levels, but this could be confounded by other factors. |
| Physical Activity Frequency (FAF) | The moderate negative correlation and downward slope emphasize the importance of physical activity in weight management. |
| Time Using Technology Devices (TUE) | The weak negative correlation and slight downward trend suggest a potential link between increased technology use and lower obesity levels, but this might be influenced by other lifestyle factors. |
| Body Mass Index (BMI) | The very strong positive correlation and clear upward trend reaffirm BMI as a highly reliable indicator of obesity. |
